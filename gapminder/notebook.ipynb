## The Evolution of Life Expectancy

```python
import plotly_express as px
import pandas as pd
```

 (Clio-Infra up to 1949; UN Population Division for 1950 to 2015) (years)

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y')

df = pd.read_csv('life-expectancy.csv', parse_dates=['Year'], date_parser=dateparse).dropna()

start_date = '1700-01-01 00:00:00'
end_date = '2015-01-01 00:00:00'

mask = (df['Year'] > start_date) & (df['Year'] <= end_date)
```

```python
px.choropleth(df, locations="Code", color="Life expectancy", hover_name="Entity", animation_frame="Year",
             color_continuous_scale=px.colors.sequential.Greens, width=800, projection="natural earth")
```
