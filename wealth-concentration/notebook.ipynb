```python
df = pd.read_csv('global-wealth.csv').fillna(0)
df.head()
```

```python
coords = []
pt = dict(
    type = 'scattergeo',
    lon = df['Long'],
    lat = df['Lat'],
    text = df['Billionaires'],
    marker = dict(
        size = df['Billionaires']*10,
        color = 'rgb(91,173,63)', # Koha green
        line = dict(width=0.5, color='rgb(40,40,40)'),
        sizemode = 'area'
    ),
    name = df['City'])
coords.append(pt)

layout = dict(
        showlegend = False,
        geo = dict(
            scope='world',
            projection=dict(type='eckert4'),
            showland = True,
            showcountries= False,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)"
        ),
    height=700,
    width=900
    )

fig = dict( data=coords, layout=layout )
iplot(fig, validate=False)
```
